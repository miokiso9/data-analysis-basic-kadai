```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# --- 左上：折れ線グラフ ---
x1 = [0, 1, 2, 3]
y1 = [1, 2, -5, 2]
axs[0, 0].plot(x1, y1)

# --- 中央上：波のグラフ ---
x2 = np.linspace(0, 10, 100)
y2 = 2 + 2 * np.sin(2 * x2)
axs[0, 1].plot(x2, y2)
axs[0, 1].set(xlim=(0, 10), ylim=(-1, 5))

# --- 右上：ヒストグラム ---
data_h = np.random.normal(15, 5, 2000) # ランダムなデータ
axs[0, 2].hist(data_h, bins=10)

# --- 左下：散布図（点のグラフ） ---
x_s = np.random.normal(3, 2, 60)
y_s = np.random.normal(4, 2, 60)
axs[1, 0].scatter(x_s, y_s)

# --- 中央下と右下は空っぽのまま ---

# グラフ同士が重ならないように調整
plt.tight_layout()
plt.show()
```
